```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path=('/content/drive/MyDrive/employee_data_250.csv')
df = pd.read_csv(file_path)
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Salary'], bins=10, kde=True)
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.title('Distribution of Salary among Employees')
plt.show()
```

```python
salary_counts, salary_bins, _ = plt.hist(df['Salary'], bins=10)
most_frequent_range = (salary_bins[salary_counts.argmax()], salary_bins[salary_counts.argmax() + 1])
print(f"Most common salary range: {most_frequent_range}")
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Experience (Years)'], y=df['Salary'], hue=df['Job Satisfaction'], palette='coolwarm')
plt.xlabel('Years of Experience')
plt.ylabel('Salary (USD)')
plt.title('Experience vs Salary')
plt.show()
```

```python
correlation = df['Experience (Years)'].corr(df['Salary'])
print(f"Correlation between Experience and Salary: {correlation:.2f}")
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x=df['Department'], order=df['Department'].value_counts().index)
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.title('Number of Employees per Department')
plt.xticks(rotation=45)
plt.show()
```

```python
top_department = df['Department'].value_counts().idxmax()
print(f"Department with the highest number of employees: {top_department}")
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Department', hue='Job Satisfaction', data=df)
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.title('Job Satisfaction Distribution Across Departments')
plt.xticks(rotation=45)
plt.legend(title='Job Satisfaction')
plt.show()
```
